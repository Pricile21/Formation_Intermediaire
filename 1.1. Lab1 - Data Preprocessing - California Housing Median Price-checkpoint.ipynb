### Etude basique de valeurs aberrantes

```python
## description du jeu de données et interpretation
# comparaison des valeurs extrêmes au valeurs centrales
# Visualiser les outliers en utilisant un boxplot avec Seaborn
plt.figure(figsize=(12, 6))
sns.boxplot(data=housing_df, x='median_house_value')
plt.xlabel('Prix des maisons')
plt.title("Boxplot des prix des maisons en californie")
plt.show()
```

```python
# Identification des données aberrantes - IQR
# Fonction pour détecter les outliers en utilisant l'IQR
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]

# Trouver les outliers pour la colonne 'median_house_value'
outliers = detect_outliers_iqr(housing_df, 'median_house_value')

# Afficher les outliers
print("Outliers détectés:")
print(outliers)
```

```python
# Identification des données aberrantes - Z-score
# Fonction pour détecter les outliers en utilisant le z-score
def detect_outliers_zscore(data, column, threshold=2):
    z_scores = (data[column] - data[column].mean()) / data[column].std()
    return data[np.abs(z_scores) > threshold]

# Trouver les outliers pour la colonne 'median_house_value'
outliers = detect_outliers_zscore(housing_df, 'median_house_value')

# Afficher les outliers
print("Outliers détectés:")
print(outliers)
```
